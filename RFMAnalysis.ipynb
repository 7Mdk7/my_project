**RFM Analysis**

```python
# prompt: rfm

import pandas as pd
import datetime as dt

# Convert InvoiceDate to datetime object
ret_df['InvoiceDate'] = pd.to_datetime(ret_df['InvoiceDate'])

# Calculate Recency, Frequency, and Monetary Value
now = dt.datetime(2011, 12, 10) # Assuming this is the current date for analysis
rfm = ret_df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (now - x.max()).days,
                                        'InvoiceNo': 'count',
                                        'totalprice': 'sum'})

# Rename columns
rfm.rename(columns={'InvoiceDate': 'Recency',
                   'InvoiceNo': 'Frequency',
                   'totalprice': 'MonetaryValue'}, inplace=True)

# Display the RFM table
print(rfm)
```

```python
rfm1=rfm[rfm['MonetaryValue']>0]
rfm1
```

```python
# prompt: label the columns , name it by a number

import pandas as pd
# Create labels for Recency, Frequency, and MonetaryValue


#r_labels = range(4, 0, -1)
#f_labels = range(1, 5)
#m_labels = range(1, 5)

# Assign labels based on quantiles
r_groups = pd.qcut(rfm1['Recency'], q=5, labels=[5,4,3,2,1])
f_groups = pd.qcut(rfm1['Frequency'], q=5, labels=[1,2,3,4,5])
m_groups = pd.qcut(rfm1['MonetaryValue'], q=5, labels=[1,2,3,4,5])

# Add RFM score to the dataframe
rfm1 = rfm1.assign(R=r_groups.values, F=f_groups.values, M=m_groups.values)

# Combine RFM score to get a single RFM score
rfm1['RFM_Score'] = rfm1['R'].astype(str) + rfm1['F'].astype(str)

rfm1
```

```python
seg_map={
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'

}
rfm1['segment']=rfm1['RFM_Score'].replace(seg_map,regex=True)
rfm1.head()
```

```python
rfm1[['segment','Recency','Frequency','MonetaryValue']].groupby('segment').agg(['mean','count','max']).round()
```
